```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

os.makedirs("figures", exist_ok=True)
os.makedirs("outputs", exist_ok=True)

sns.set(style="whitegrid", context="talk")
np.random.seed(42)
```

```python
data = [
    ["PIVC ROI definition", "roi", 0.82],
    ["TPJ boundary choice", "roi", 0.74],
    ["Motion correction", "preprocessing", 0.91],
    ["Global signal regression", "preprocessing", 0.59],
    ["Connectivity estimator", "model", 0.77],
    ["Thresholding strategy", "model", 0.63],
    ["Parcellation granularity", "roi", 0.71],
    ["Multiple comparison control", "statistics", 0.80]
]

df = pd.DataFrame(
    data,
    columns=["decision", "stage", "impact_score"]
)

df
```

```python
summary = df.sort_values("impact_score", ascending=False)
summary
```

```python
plt.figure(figsize=(12,5))

ax = sns.barplot(
    data=summary,
    x="decision",
    y="impact_score",
    hue="stage"
)

plt.xticks(rotation=35, ha="right")
plt.xlabel("")
plt.ylabel("Relative Impact Score")
plt.title("Vestibular Reproducibility Decision Landscape")

ax.legend(
    title="stage",
    fontsize=10,
    title_fontsize=10,
    bbox_to_anchor=(1.02, 1),
    loc="upper left",
    borderaxespad=0
)

plt.tight_layout()
plt.savefig("figures/01_roi_pipeline_landscape.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
df.to_csv("outputs/01_roi_pipeline_landscape.csv", index=False)
summary.to_csv("outputs/01_roi_pipeline_summary.csv", index=False)

summary
```
